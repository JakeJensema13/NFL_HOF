```python
WR_data = df.loc[df.position == 'WR']

fig = px.bar(WR_data, x = 'player', y = 'years_as_starter',color = 'years_as_starter', title = 'HOF WR Years Starting')

fig.update_layout(template = 'plotly_dark', xaxis_title = 'Player Name', yaxis_title = 'Years As Starter')

fig.show()



```

```python
RB_Data = df.loc[df.position == 'RB']

fig = px.scatter(RB_Data, x = 'player', y = 'longest_rush', color = 'player')


fig.show()
```

```python
df['Years_Active'] = df['to'] - df['from']

fig = px.box(df, x='position' ,y='Years_Active', color = 'position', title = 'Years active per position')

fig.update_layout(template = 'plotly_dark', font = dict(family = 'Arial', size = 16))

fig.show()
```

```python
df.loc[df.Years_Active > 20]
```

```python
fig = px.pie(df, values = 'pro_bowl_selections', names= 'position', title = 'Total Pro Bowl Selections By Poition', hover_data = ['all_pro_selections'])
fig.show()
```

```python
QB_data = df.loc[df.position == 'QB']
sortedQB_data = QB_data.sort_values(by='pass_td')
fig = px.bar(sortedQB_data, x= 'player', y= 'pass_td', color= 'pass_td', title = 'HOF QB Total TD Thrown')

fig.update_layout(yaxis_title = 'TD Thrown', template = 'plotly_dark', font = dict(family = 'Verdana', size = 13))
fig.show()
```

```python
sortedQB_data['QB_TD'] = df['pass_td'] / (df['to'] - df['from'])

fig = px.bar(sortedQB_data, x='player', y='QB_TD', title = 'Average QB Passing TD Per Year')

fig.update_layout(template = 'ggplot2')
fig.show()
```

```python
WR_RB_Data = df.loc[(df.position == 'RB') | (df.position == 'WR')]
WR_RB_Data['AVG_TD'] = df['td_total'] / (df['to'] - df['from'])
WR_RB_Data_Sorted = WR_RB_Data.sort_values(by='AVG_TD')




fig = px.bar(WR_RB_Data_Sorted, x='player', y= 'AVG_TD')
fig.update_layout(template = 'plotly_dark')
fig.show()

```
